```python
import csv
import json

# 地名を読み込み
place_names = []
places = []

with open('latest.csv', 'r', encoding='utf-8') as f:
    for row in csv.DictReader(f):
        place = {}
        place['name'] = row['市区町村名'] + row['大字町丁目名'] + row['小字・通称名']
        place['kana'] = kata_to_hira(row['市区町村名カナ'] + ' ' +row['大字町丁目名カナ'])
        place['lat'] = row['緯度']
        place['lon'] = row['経度']
        place_names.append(row)
        places.append(place)

# アイドル名を読み込む
idol_names = []
with open('idol_names.csv', 'r', encoding='shift_jis') as f:
    for row in csv.DictReader(f):
        idol_names.append(row)
```

```python
# https://teratail.com/questions/153806
def kata_to_hira(strj):
    return "".join([chr(ord(ch) - 96) if ("ァ" <= ch <= "ヴ") else ch for ch in strj])
```

```python
kata_to_hira('カナ')
```

```python
idol_places = {}
for idol in idol_names:
    idol_places[idol['名前']] = []
    for key, name in idol.items():
        for place in places:
            #　名前が空の場合の処理
            if len(name) < 1:
                continue
            if name in place['name']:
                idol_places[idol['名前']].append(place)
                continue
            if name in place['kana']:
                idol_places[idol['名前']].append(place)
                continue
```

```python
for idol in idol_places:
    print(idol, len(idol_places[idol]))
```

```python
with open('idol_places.json', mode='wt', encoding='utf-8') as file:
  json.dump(idol_places, file, ensure_ascii=False, indent=2)
```
